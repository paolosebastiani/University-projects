# Loading Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

# **Baseline Model - Linear Approximation**

## 1. Define Model Parameters



```python
# Simulation parameters
N_AGENTS = 10000
N_RUNS = 5000

# Agent preferences (intrinsic payoffs)
A_R, B_R = 1.2, 1.0  # Payoffs for R-agents (prefer A)
A_S, B_S = 1.0, 1.2  # Payoffs for S-agents (prefer B)

# Technology Learning Parameters
# Technology A
L_A = 10.0  # Max learning potential
K_A = 0.01 # Steepness of learning
N0_A = 1000 # Adoption midpoint (10% of total agents)
# Technology B
L_B = 10.0
K_B = 0.01
N0_B = 1000

# List to store the final market share of Technology A from each run
final_market_shares = []
```

## 2. Loops

```python
# --- The Outer Loop (Monte Carlo Runs) ---
for run in range(N_RUNS):
    # Initialize state for this single run
    n_A, n_B = 0, 0

    # Create a shuffled sequence of agents (the "small historical events")
    agent_sequence = ['R'] * (N_AGENTS // 2) + ['S'] * (N_AGENTS // 2)
    random.shuffle(agent_sequence)

    # --- The Inner Loop (Single Adoption History) ---
    for agent_type in agent_sequence:
        # Calculate payoffs based on the current state (n_A, n_B)
        payoff_A = (A_R if agent_type == 'R' else A_S) + (L_A / (1 + np.exp(-K_A * (n_A - N0_A))))
        payoff_B = (B_R if agent_type == 'R' else B_S) + (L_B / (1 + np.exp(-K_B * (n_B - N0_B))))

        # Agent makes a choice
        if payoff_A >= payoff_B:
            n_A += 1
        else:
            n_B += 1

    # After the run is complete, store the final market share
    final_market_shares.append(n_A / N_AGENTS)
```

## 3. Visualize Results

```python
# --- 4. Analyze and Visualize Results ---
# At the end of all runs, create a histogram of the outcomes
plt.figure(figsize=(10, 6))
plt.hist(final_market_shares, bins=25, edgecolor='black')
plt.title(f'Distribution of Final Market Shares for Tech A over {N_RUNS} runs')
plt.xlabel('Final Market Share of Technology A')
plt.ylabel('Frequency (Number of Runs)')
plt.xlim(0, 1) # Market share is between 0 and 1
plt.grid(axis='y', alpha=0.75)
plt.show()
```
